Load in the input text file and store each line as an int in a list

```python
with open("input.txt") as input_file:    
    input_list = [item.strip('\n') for item in input_file.readlines()]
```

Print the input so we can check that it looks correct

```python
input_list
```

Found this command on the internet:
We want a new list group every time there is an empty entry in the list. 'groupby' provides a new group everytime the key changes. Groupby outputs an iterable, where k is when the value of the key changes, and g are the values contained within each group (once casted to a list). The 'if k' part is to say, only keep the groups containing values when the key changes to positive, i.e. non-empty groups.

When set to bool, this means everytime we get an empty group, 

```python
from itertools import groupby
seperated_list = [list(g) for k, g in groupby(input_list, key=bool) if k]
print(seperated_list)
```

Now we want to sum up the contents of each seperated list. We use a comprehension to avoid using a for loop. Map allows us to apply a function to each item in a list, returning an iterable. We convert this iterable back to a list to get a list of integers and then sum each list to get a single value for each list in our seperated list


```python
summed_list = [sum(list(map(int, list_of_list))) for list_of_list in seperated_list]
summed_list
```

Now the answer is just the maximum value found in our summed sub-answers 

```python
answer1 = max(summed_list)
answer1
```

To find the sum of the top 3 most calories we want to sort the list by values (sort alters the values in place, where as sorted does not)

```python
summed_list.sort()
summed_list
```

Now we just need to sum the last 3 entries in our ascended sort of values. If we use [-3:] this means take the numbers located at: 3 back from the end up to the very last number. If we had specified [0:] this would mean from the first up to the end (so the minus means just count back from the end and the : means the index of the last +1)

```python
answer2 = sum(summed_list[-3:])
answer2
```
