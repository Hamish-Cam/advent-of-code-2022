We need to calculate the total score of all the rock, paper, scissors decisions. Start by reading in the data

```python
with open("input.txt") as input_file:    
    input_list = [item.strip('\n').split() for item in input_file.readlines()]
    
input_list
```

Define the payoffs for each possible set of circumstances:

```python
score_dict = {
  'A': {'X': 4,
        'Y': 8,
        'Z': 3,},
  'B': {'X': 1,
        'Y': 5,
        'Z': 9,},
  'C': {'X': 7,
        'Y': 2,
        'Z': 6,},
}
```

Now we need to go through each pair of actions and get the payoff for each game

```python
payoffs = [score_dict[item[0]][item[1]] for item in input_list]
payoffs
```

The total score is then the sum of the payoffs

```python
answer1 = sum(payoffs)
answer1
```
